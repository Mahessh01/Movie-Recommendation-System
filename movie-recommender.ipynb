```python
data = data[['genres', 'id', 'keywords', 'title', 'overview', 'cast','crew']]
```

```python
import ast
def convert(obj):
    l= []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data.head(1)
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
data.head(1)
```

```python
import ast
def convert2(obj):
    l= []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
data['cast']=data['cast'].apply(convert2)
```

```python
data.head(1)
```

```python
import ast

def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew']=data['crew'].apply(fetch_director)
```

```python
data.head(4)
```

```python
data['overview'][0]
```

```python
data['overview']=data['overview'].apply(lambda x:x.split())
```

```python
data.head(2)
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast']=data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head(2)
```

```python
data['tags'] = data['overview']+data['genres']+data['keywords']+data['cast']+data['crew']
```

```python
data.head(2)
```

```python
df = data[['id','title','tags']]
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
print(type(data['tags'].iloc[0]))
print(data['tags'].iloc[0])
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
print(cv.get_feature_names_out()[:200])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse= True,key= lambda x:x[1])[1:6]
```

```python
#main function
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances= similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse= True,key= lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
     
```

```python
recommend('Batman Begins')
```
